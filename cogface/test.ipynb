```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook

with open(r"database.pkl", mode="rb") as db:
    database = pickle.load(db)

input = database["Hyojae"].descriptors + database["Jacob"].descriptors + database["Joseph"].descriptors + database["Kathryn"].descriptors + database["Dharshini"].descriptors[0:4] + database["Angelina"].descriptors[0:4]


graph, adj_mat, cluster_count = whispers(input, 0.5)
for i in range(len(input)):
    print("ID: ", graph[i].id, " LABEL: ", graph[i].label)

fig, ax = plt.subplots()

ax.plot(np.arange(len(cluster_count)), cluster_count)
ax.grid()

plot_graph(graph, adj_mat)
```
